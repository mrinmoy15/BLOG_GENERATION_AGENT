```python
from datetime import date
from langchain_openai import ChatOpenAI
from src.core.main_graph import main_graph

# 1. Build the graph
graph = main_graph()

# 2. Define the LLM
llm = ChatOpenAI(model="gpt-4o", temperature=0.3)

# 3. Initial state — only topic and recency fields are needed from the user.
#    Everything else gets populated by the nodes as the graph runs.
initial_state = {
    "topic": "How LLM inference optimization works: KV cache, speculative decoding, and batching",
    "as_of": date.today().isoformat(),  # e.g. "2026-03-26"
    "recency_days": 0,   # router will overwrite this — safe placeholder
    # --- everything below is populated by nodes, but TypedDict needs them declared ---
    "mode": "closed_book",       # router overwrites
    "needs_research": False,     # router overwrites
    "queries": [],               # router overwrites
    "evidence": [],              # research node overwrites (if needed)
    "plan": None,                # orchestrator overwrites
    "sections": [],              # workers populate via operator.add
    "merged_md": "",             # merge_content populates
    "md_with_placeholders": "",  # decide_images populates
    "image_specs": [],           # decide_images populates
    "final": "",                 # generate_and_place_images populates
}

# 4. Config carrying the LLM
config = {"configurable": {"llm": llm}}

# 5. Invoke
result = graph.invoke(initial_state, config=config)
```
